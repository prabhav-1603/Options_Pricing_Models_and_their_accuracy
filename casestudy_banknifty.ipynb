```python
import pandas as pd

# Load historical data and options data
historical_data = pd.read_csv('datasets/Historical_data.csv')
options_data = pd.read_csv('datasets/new_options_data.csv')
options_data_put = pd.read_csv('datasets/new_options_data_put.csv')

# Remove blank space from column names in options data
options_data.columns = [col.strip() for col in options_data.columns]
options_data_put.columns = [col.strip() for col in options_data_put.columns]

# Remove rows with values=nan(missing values) from historical data
historical_data = historical_data.dropna()

# Select relevant columns from options data and drop rows with values=nan(missing values)
options_data = options_data[['Symbol', 'Date', 'Expiry', 'Option type', 'Strike Price', 'Close']].dropna()

# Convert Expiry and Date columns to datetime format
options_data['Expiry'] = pd.to_datetime(options_data['Expiry'], format='%d%b%Y')
options_data['Date'] = pd.to_datetime(options_data['Date'], format='%d%b%Y')

# Select relevant columns from put options data and drop rows with missing values
options_data_put = options_data_put[['Symbol', 'Date', 'Expiry', 'Option type', 'Strike Price', 'Close']].dropna()

# Convert Expiry  and Date columns to datetime format for put options data
options_data_put['Expiry'] = pd.to_datetime(options_data_put['Expiry'], format='%d%b%Y')
options_data_put['Date'] = pd.to_datetime(options_data_put['Date'], format='%d%b%Y')

# Print column names for debugging
#print(options_data_put.columns)
#print(options_data.columns)

# Concatenate(Merge) options data and put options data
options_data = pd.concat([options_data, options_data_put])

# Sort options data by Date and reset index
options_data = options_data.sort_values(by='Date').reset_index(drop=True)

# Display the final combined options data
print(options_data)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import norm

log_returns = np.log(historical_data["Close"] / historical_data["Close"].shift(1))
annual_volatility = log_returns.std() * np.sqrt(252)  # Annualizing the standard deviation
print(annual_volatility)
# Define Black-Scholes model
def black_scholes(S, K, T, r, sigma, option_type='CE'):
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    if option_type == 'CE':
        price = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    else:
        price = K * np.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
    return price

# Define Binomial model
import numpy as np

def binomial_model(S, K, T, r, sigma, steps=10, option_type='CE'):
    dt = T / steps
    u = np.exp(sigma * np.sqrt(dt))
    d = 1 / u
    p = (np.exp(r * dt) - d) / (u - d)
    discount_factor = np.exp(-r * dt)
    
    # Initialize asset prices at maturity
    prices = np.zeros(steps + 1)
    for i in range(steps + 1):
        prices[i] = S * (u ** (steps - i)) * (d ** i)
    
    # Initialize option values at maturity
    if option_type == 'CE':
        option_values = np.maximum(prices - K, 0)
    else:
        option_values = np.maximum(K - prices, 0)
    
    # Backward induction for option price
    for i in range(steps - 1, -1, -1):
        option_values = discount_factor * (p * option_values[:-1] + (1 - p) * option_values[1:])
        
    return option_values[0]


# Analysis parameters
r = 0.0676 # Risk-free rate(saw from government bonds data from internet around the 1 month period we are taking into account)
current_index_price = historical_data["Close"].iloc[-1]  # Most recent index price

# Comparing model prices to market prices
results=[]

for index, row in options_data.iterrows():
    S = current_index_price
    K = row['Strike Price']
    T = (row['Expiry'] - row['Date']).days / 365.0
    if T == 0:
        print(row['Date'])
    market_price = row['Close']
    option_type = row['Option type']

    bs_price = black_scholes(S, K, T, r, annual_volatility, option_type)
    bin_price = binomial_model(S, K, T, r, annual_volatility, option_type=option_type)

    results.append({
        'Type': option_type,
        'Strike': K,
        'Expiry': row['Expiry'],
        'Market Price': market_price,
        'BS Price': bs_price,
        'Binomial Price': bin_price,
        'BS Error': abs(bs_price - market_price),
        'Binomial Error': abs(bin_price - market_price),
        'Date': row['Date'],
        '% Binomial Error': (bin_price-market_price)/market_price,
        '% BS Error': (bs_price-market_price)/market_price
    })

results_df = pd.DataFrame(results)

print(results_df)

# Summary statistics
print("Mean Error Black-Scholes: ", results_df['% BS Error'].mean())
print("Mean Error Binomial: ", results_df['% Binomial Error'].mean())
```

```python
def backtest(options_data, model_name='BS'):
    initial_cash = 10000  # starting capital
    cash = initial_cash
    positions = []
    transaction_log = []

    for index, row in options_data.iterrows():
        model_price = row[f'{model_name} Price']
        market_price = row['Market Price']
        strike_price = row['Strike']
        option_type = row['Type']
        expiry = row['Expiry']
        date = row['Date']

        # Decision making
        if model_price > market_price*1.05:  # Buying if model thinks it's undervalued
            if cash >= market_price:
                positions.append({'purchase_price': market_price, 'strike_price': strike_price, 'expiry': expiry, 'type': option_type})
                cash -= market_price
                transaction_log.append(f"Bought {option_type} for ${market_price} on {date}")

        # Closing/Selling positions
        for position in positions:
            if date >= position['expiry'] or model_price < market_price * 0.95:  # Selling if model thinks it's overvalued or at expiry
                cash += market_price
                positions.remove(position)
                transaction_log.append(f"Sold {position['type']} for ${market_price} on {date}")

    final_cash = cash + sum(pos['purchase_price'] for pos in positions)  # Liquidate all positions
    profit_loss = final_cash - initial_cash

    return profit_loss, transaction_log

# Backtest for both models
profit_loss_bs, transactions_bs = backtest(results_df, 'BS')
profit_loss_binomial, transactions_binomial = backtest(results_df, 'Binomial')

print("Profit/Loss for Black-Scholes Model:", profit_loss_bs)
# print("Transactions for Black-Scholes Model:", transactions_bs)
print("\nProfit/Loss for Binomial Model:", profit_loss_binomial)
# print("Transactions for Binomial Model:", transactions_binomial)
```
